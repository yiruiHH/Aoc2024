```python
input = open('input.txt', 'r')
left = []
right = []
for line in input.readlines():
    line = line.replace('\n', '')
    numbers = line.split('   ')
    left.append(int(numbers[0]))
    right.append(int(numbers[1]))
left.sort()
right.sort()
distance = 0
for i in range(len(left)):
    distance += abs(left[i] - right[i])
print(distance)
```

```python
rd = {}
for r in right:
    if (rd.get(r) is None):
        rd.update({r:1})
    else:
        rd.update({r: rd.get(r) + 1})
print(rd)
similarity_score = 0
for l in left:
    if (rd.get(l) is not None):
        similarity_score += l * rd.get(l)
print(similarity_score)
```
